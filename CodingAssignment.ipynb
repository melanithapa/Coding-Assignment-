```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
IMDB_df = pd.read_csv("IMDB_top_1000.csv")

# Take the top 20 movies
top_20_movies = IMDB_df.head(20)

Rate = top_20_movies['Rate']
Genre = top_20_movies['Genre']

# Create a bar graph comparing ratings by genre
plt.figure(figsize=(12, 6))
plt.barh(Genre, Rate, color='g')
plt.xlabel('Rating')
plt.ylabel('Genre')
plt.title('Top 20 Movies: Ratings by Genre')
plt.tight_layout()

# Show the graph
plt.show()
```

The graph displays the top 20 movies from the IMDB dataset and illustrates how movie ratings are distributed across various genres. This visualization is helpful for identifying which genres tend to have higher-rated movies among the top 20 in the dataset. Additionally, I noticed that genre drama was the highest rated and there are many combinations with drama.
